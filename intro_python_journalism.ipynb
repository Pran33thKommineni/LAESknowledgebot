## 2. Variables

Variables are names that hold values. You assign with `=`. Think of them like labels on boxes—you put something in the box and can use the label to get it back.

```python
# Create a variable and use it
headline = "City Council Approves New Budget"
print(headline)
```

Variables can be reused and updated. Use clear names so your code is easy to read (and so you remember what you meant when you come back to a story).

```python
reporter_name = "Alex"
article_count = 5

print("Reporter:", reporter_name)
print("Articles this month:", article_count)
```

## 3. Common Data Types

Python has a few types you'll use all the time:

| Type | What it is | Example |
|------|------------|--------|
| **str** (string) | Text in quotes | `"Breaking news"` |
| **int** (integer) | Whole numbers | `42`, `1000` |
| **float** | Numbers with decimals | `3.14`, `99.5` |
| **bool** (boolean) | True or False | `True`, `False` |

### Strings (text)

Strings are for words, sentences, headlines—any text. Use single or double quotes.

```python
quote = "The first draft is just you telling yourself the story."
source = 'Stephen King'

print(quote)
print("—", source)
print("Type of quote:", type(quote))
```

### Integers and floats (numbers)

Use **integers** for counts (votes, population, number of sources). Use **floats** when you need decimals (percentages, averages, rates).

```python
votes_for = 12500
votes_against = 8300
turnout_percent = 67.3

print("Votes for:", votes_for, "|", type(votes_for))
print("Votes against:", votes_against)
print("Turnout:", turnout_percent, "%", "|", type(turnout_percent))
```

### Booleans (True / False)

Booleans answer yes/no questions. Useful for fact-checking flags, “was this verified?”, or simple conditions.

```python
story_published = True
fact_checked = False

print("Story published?", story_published)
print("Fact checked?", fact_checked)
print("Type:", type(story_published))
```

### Quick practice: mix types with f-strings

You can combine variables into one message using **f-strings**: put `f` before the quote and use `{variable_name}` inside.

```python
city = "Springfield"
population = 115000
mayor_approval = 58.2

sentence = f"In {city}, population {population}, mayor approval is {mayor_approval}%."
print(sentence)
```

## 4. Lists

A **list** holds multiple values in order. Use square brackets `[]` and separate items with commas. Great for sources, headlines, data points, or any collection you need to work with.

### Creating and viewing lists

```python
# A list of sources for a story
sources = ["City Hall spokesperson", "Council member Chen", "Local advocacy group"]

# A list of headlines (or numbers—lists can hold any type)
daily_reads = [1200, 3400, 890, 2100]

print("Sources:", sources)
print("Reads per day:", daily_reads)
print("Type:", type(sources))
```

### Accessing elements (indexing)

Each item has a position, or **index**. Indexes start at **0**: first item is `0`, second is `1`, and so on. Use `list[index]` to get one item. Use **negative** indexes to count from the end: `-1` is the last item.

```python
sources = ["City Hall spokesperson", "Council member Chen", "Local advocacy group"]

print("First source (index 0):", sources[0])
print("Second source (index 1):", sources[1])
print("Last source (index -1):", sources[-1])
```

### Useful list operations

- **`len(list)`** — number of items in the list  
- **`list.append(item)`** — add one item to the end  
- You can change an item: **`list[index] = new_value`**

```python
headlines = ["Budget passes", "Mayor responds", "Protest planned"]

print("Number of headlines:", len(headlines))

headlines.append("Update: vote delayed")
print("After adding one more:", headlines)
```
